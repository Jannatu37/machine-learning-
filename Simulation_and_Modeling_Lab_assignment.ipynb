```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
vector1 = np.random.rand(15)
vector2 = np.random.rand(15)
print("vector1:",vector1)
print("vector2:",vector2)

plt.figure(figsize=(10, 10))
plt.plot(vector1, label="vector1", marker='*' , color = "Red")
plt.plot(vector2, label="Vector 2", marker='D' , color = "Black")

plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Plot of Two Random Vectors")
plt.legend()
plt.grid(True)
plt.show()
```

```python
matrix = np.random.rand(4, 4)
print(matrix)
plt.figure(figsize=(10, 5))
sns.heatmap(matrix, annot=True, cmap='twilight', xticklabels=['1', '2', '3', '4'], yticklabels=['Alpha', 'Beta', 'Ghama', 'Omega'])
plt.title('4x4 Matrix Heatmap')
plt.show()
```

```python

matrix1 = np.random.rand(4, 4)
matrix2 = np.random.rand(4, 4)
print("Matrix 1:\n", matrix1)
print("Matrix 2:\n", matrix2)

add = matrix1 + matrix2
sub = matrix1 - matrix2
multi = np.dot(matrix1, matrix2)

print("Addition Result:\n", add)
print("Subtraction Result:\n", sub)
print("Multiplication Result:\n", multi)

fig, axes = plt.subplots(1, 3, figsize=(9, 5))

axes[0].bar(range(16), add.flatten(), color='green')
axes[0].set_title('Matrix Addition')
axes[0].set_xlabel('Element Index')
axes[0].set_ylabel('Value')

axes[1].bar(range(16), sub.flatten(), color='Purple')
axes[1].set_title('Matrix Subtraction')
axes[1].set_xlabel('Element Index')
axes[1].set_ylabel('Value')


axes[2].bar(range(16), multi.flatten(), color='gray')
axes[2].set_title('Matrix Multiplication')
axes[2].set_xlabel('Element Index')
axes[2].set_ylabel('Value')

plt.tight_layout()
plt.show()
```
